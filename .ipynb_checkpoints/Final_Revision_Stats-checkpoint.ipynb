```python
import pandas as pd
import numpy as np 
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.proportion import proportions_ztest
```

จากการสุ่มผลไม้กระป๋อง จํานวน 10 กระป๋อง ที่ผลิตจากโรงงานหนึ่งได้นํ้าหนักสุทธิเป็น 164 161 158 170 161 152 158 169 152
และ 160 กรัม ทางโรงงานได้โฆษณาว่านํ้าหนักเฉลี่ยของผลไม้กระป๋องมีค่าเท่ากับ 160 กรัม จงทดสอบว่าคําโฆษณาของโรงงานนี้เป็นจริง
หรือไม่กําหนดระดับนัยสําคัญ 0.05

ต้องการทดสอบว่านํ้าหนักเฉลี่ยของผลไม้กระป๋องมีค่าเท่ากับ 160 กรัม ตามที่โรงงานกล่าวอ้างหรือไม่
กําหนด u คือนํ้าหนักเฉลี่ยของผลไม้กระป๋อง

```python
# H0 : u  = 160
# H1 : u != 160
# u แทนค่าน้ำหนักเฉลี่ยของผลไม้กระป๋อง
alpha = 0.05
data = [164,161,158,170,161,152,158,169,152,160]
xBar = np.mean(data)
s = np.std(data, ddof=1)
tCal = (xBar - 160) / (s / np.sqrt(len(data)))
t_crit_left = stats.t.ppf(alpha/2,df=9)
t_crit_right = stats.t.isf(alpha/2,df=9)
pVal = 2*stats.t.sf(tCal,df=9)
print(t_crit_left , tCal , t_crit_right , pVal)
```

พรรคการเมืองหนึ่งทําการสํารวจความนิยมของประชาชนที่มีต่อพรรคโดยสุ่มตัวอย่างมา 100 คน ปรากฏว่ามีผู้นิยมพรรคการเมือง
ดังกล่าว 60 คน ต้องการทดสอบว่าในการเลือกตั้งครั้งต่อไปผู้สมัครของพรรคการเมืองนี้จะมีโอกาสได้รับเลือกมากกว่า 50% หรือไม่
กําหนดระดับนัยสําคัญ 0.05

```python
# H0 : p <= 0.5
# H1 : p > 0.5
# p แทน สัดส่วนของผู้นิยมพรรคดังกล่าว
pHat = 60/100
alpha = 0.05
zCal = ( pHat - 0.5) / np.sqrt((0.5 * 0.5) / 100)
z_crit = stats.norm.isf(alpha)
pVal = stats.norm.sf(zCal)
zCal , z_crit , pVal
```

จากการทดสอบถ่าน AAA 1.5V 2 ยี่ห้อคือ A และ B ว่ามีอายุการใช้งานแตกต่างกันหรือไม่ ในการทดสอบใช้ถ่าน AAA ยี่ห้อละ 10 ก้อน
แต่ละก้อนนําไปใช้กับเครื่องมืออิเล็กทรอนิกส์ชนิดหนึ่งโดยเปิดใช้งานติดต่อกันจนกว่าเครื่องมือจะหยุดทํางานและทําการบันทึกระยะเวลา
ตั้งแต่เริ่มจนหยุดการทํางาน ได้ข้อมูลดังนี้
ยี่ห้อ A: 5.47, 3.91, 5.81, 4.11, 6.86, 4.52, 5.24, 4.96, 5.64, 6.73
ยี่ห้อ B: 6.04, 7.93, 7.53, 7.69, 6.90 , 6.68, 6.00, 7.63, 5.86, 8.26
จงทดสอบว่าอายุการใช้งานของถ่าน 2 ยี่ห้อแตกต่างกันหรือไม่ที่ระดับนัยสําคัญ 0.05 ถ้าถือว่าอายุการใช้งานของถ่านทั้ง 2 ยี่ห้อ มีความ
แปรปรวนเท่ากัน 

```python
# H0 : u1 - u2  = 0
# H0 : u1 - u2 != 0
# u1 แทนอายุการใช้งานของถ่าน ยี่ห้อ A
# u2 แทนอายุการใช้งานของถ่าน ยี่ห้อ B
# ความแปรปรวนเท่ากัน
alpha = 0.05
A = [5.47, 3.91, 5.81, 4.11, 6.86, 4.52, 5.24, 4.96, 5.64, 6.73]
B = [6.04, 7.93, 7.53, 7.69, 6.90 , 6.68, 6.00, 7.63, 5.86, 8.26]
tCal , pVal = stats.ttest_ind(a=A,b=B,equal_var=True,alternative="two-sided")
t_crit_left = stats.t.ppf(alpha/2 , df = 18)
t_crit_right = stats.t.isf(alpha/2 , df = 18)
print(t_crit_left , tCal , t_crit_right)
```

เกษตรกรผู้หนึ่งทดลองปลูกถั่วเหลือง 2 พันธุ์พันธุ์ที่ 1 ปลูกในแปลงทดลอง 8 แปลง พันธุ์ที่2 ปลูกในแปลงทดลอง 10 แปลง ได้ผลผลิต
ถังต่อไร่ดังนี้
พันธ์ที่ 1: 36, 34, 32, 35, 50, 46, 45, 38
พันธ์ที่2: 35, 39, 37, 38, 35, 42, 43, 36, 39, 40
ถ้าผลผลิตของถั่วเหลืองมีการแจกแจงแบบปกติจงทดสอบว่าความแปรปรวนของผลผลิต ถั่วเหลืองทั้งสองพันธุ์นี้แตกต่างกันหรือไม่ ที่
ระดับนัยสําคัญ 0.01

```python
# omega1^2 / omega2^2  = 1
# omega1^2 / omega2^2 != 1
# omega คือ ความแปรปรวนของผลผลิต
A = [36, 34, 32, 35, 50, 46, 45, 38]
SA = np.std(A ,ddof=1)
B = [35, 39, 37, 38, 35, 42, 43, 36, 39, 40]
SB = np.std(B ,ddof=1)
alpha = 0.01
fCal = (SA**2)/(SB**2)
f_crit_left = stats.f.ppf(alpha/2 , dfn=7 , dfd=9)
f_crit_right = stats.f.isf(alpha/2 , dfn=7 , dfd=9)
pVal = 2*min(stats.f.sf(fCal, dfn=7 , dfd=9),stats.f.cdf(fCal, dfn=7 , dfd=9))
print(fCal , f_crit_left , f_crit_right , pVal)
```

บริษัทผลิตหลอดไฟ LED รายย่อยแห่งหนึ่งอ้างว่าหลอดไฟที่ผลิตมีอายุการใช้งานนานกว่า 5.85 ปีจากการเก็บข้อมูล
อายุการใช้งานของหลอดไฟที่ผลิตจากบริษัทแห่งนี้จากนิคมอุตสาหกรรมทีใช้หลอดไฟมาไม่ตํ่ากว่า 7 ปีพบว่า
หลอดไฟ 30 หลอดมีอายุการใช้งานดังนี้
4.17, 5.72, 6.23, 5.65, 6.59, 6.47, 4.67, 6.25, 3.85, 7.02, 5.24
4.69, 5.75, 5.05, 5.33, 5.45, 6.32, 4.7 , 5.63, 6.22, 6.73, 4.81
4.29, 4.68, 6.13, 5.44, 6.06, 7.54, 7.28, 5.18
หากอายุการใช้งานหลอดไฟเป็นการแจกแจงปกติจงทดสอบว่าอายุการใช้งานของหลอดไฟเป็นไปตามที่บริษัทกล่าว
อ้างหรือไม่ที่ระดับนัยสําคัญ 0.05

```python
# H0 : u <= 5.85
# H1 : u > 5.85
# u แทน อายุการใช้งานของหลอดไฟ
data = [4.17, 5.72, 6.23, 5.65, 6.59, 6.47, 4.67, 6.25, 3.85, 7.02, 5.24 ,4.69, 5.75, 5.05, 5.33, 5.45, 6.32, 4.7 , 5.63, 6.22, 6.73, 4.81, 4.29, 4.68, 6.13, 5.44, 6.06, 7.54, 7.28, 5.18]
alpha = 0.05
xBar = np.mean(data)
s = np.std(data,ddof=1)

zCal = (xBar - 5.85) / (s/np.sqrt(len(data)))
print(zCal)
z_crit = stats.norm.isf(alpha)
print(ztest(x1=data,value=5.85,alternative="larger"))
print(z_crit)
```

จากการสุ่มคนมาตรวจเลือด 22 คน พบว่าค่าเฉลี่ยของกรดยูริกเป็น 4.6 และความแปรปรวน 1.44 จงทดสอบว่า ความแปรปรวนของกรด
ยูริกมีค่าแตกต่างจาก 1 หรือไม่ที่ระดับนัยสําคัญ 0.10

```python
# H0 : omega^2 = 1
# H1 : omega^2 != 1
alpha = 0.1
n = 22
xBar = 4.6
v = 1.44

chi2Cal = ((n-1)* v)/ 1 
chi2_crit_left = stats.chi2.ppf(alpha/2,df=21)
chi2_crit_right = stats.chi2.isf(alpha/2,df=21)
pVal = 2*min(stats.chi2.sf(abs(chi2Cal),df=21),stats.chi2.cdf(abs(chi2Cal),df=21))
chi2Cal , chi2_crit_left , chi2_crit_right , pVal

```

โรงงานผลิตนํ้าตาลทรายแห่งหนึ่งกล่าวว่า ปริมาณนํ้าตาลทรายบรรจุถุงขนาด 1 กิโลกรัม มีการแจกแจงแบบปกติโดยมีส่วนเบี่ยงเบน
มาตรฐานมากกว่า 30 กรัม จึงทําการสุ่มตัวอย่างนํ้าตาลทรายมาจํานวน 20 ถุง คํานวณส่วนเบี่ยงเบนมาตรฐานได้ 32 กรัม จงสรุปผลที่
ระดับนัยสําคัญ 0.01

```python
# H0 : omega <= 30 
# H1 : omega > 30
# omega แทน ส่วนเบี่ยงเบนมาตรฐานปริมาณน้ำตาลทรายบรรจุถุง

n = 20
S = 32
alpha = 0.01
chi2_cal = (n-1) * (S**2) / (30**2)
chi2_crit = stats.chi2.isf(alpha,df=n-1)
pVal = stats.chi2.sf(chi2_cal , df=n-1)
chi2_cal,  chi2_crit , pVal
```

เกษตรกรผสมเกสรถั่วชนิดเมล็ดเรียบกับถั่วชนิดเมล็ดขรุขระ ได้ถั่วรุ่นลูก F1 เป็นถั่วพันธุ์ผสม 100 ต้น นําไปปลูกให้ผลผลิตเป็นถั่วเมล็ด
เรียบ 69 ต้น และถั่วเมล็ดขรุขระ 31 ต้น จงทดสอบว่าอัตราส่วนระหว่างถั่วเมล็ดเรียบต่อถั่วเมล็ดขรุขระจะมีค่าแตกต่างจากอัตราส่วน
3:1 ตามกฎพันธุศาสตร์ของเมนเดลหรือไม่ ที่ระดับนัยสําคัญ 0.05

```python
# H0 : p1-p2 = 3:1
# H1 : p1-p2 != 3:1
O = np.array([69,31])
p = np.array([3/4,1/4])
E = 100 * p
chi2_crit = stats.chi2.isf(0.05,df=1)
print(stats.chisquare(f_obs=O,f_exp=E))
print(chi2_crit)
```
